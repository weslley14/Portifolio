###Questão 1

Importe a planilha `dados_starwars.xlsx` na pasta virtual `content`.

```python
# Questao 1
#Importando o Pandas e o Arquivo excel
import pandas as pd

dados_starwars = pd.read_excel("/content/dados_starwars.xlsx")
dados_starwars
```

###Questão 2

Remova todos os valores faltantes. Salve a base de dados atualizada.

```python
# Questao 2
# Excluindo os arquivos sem dados(NaN) e adcionando a um novo DataFrame
dados_starwars_at = dados_starwars.dropna()
print(dados_starwars_at.shape)
```

###Questão 3

Filtre somente os registros com `especie` = `humano`. Salve a base de dados atualizada.

```python
# Questao 3
#Filtrando os dados onde a especíe é igual a Humanos
dados_filtrados = dados_starwars_at[dados_starwars_at['especie'] == 'Humano']
dados_filtrados
```

###Questão 4

Edite a variável `altura` (que esta em centímetros) para que passe para a unidade em metros (*dividir por 100*). Salve a base de dados atualizada.

```python
# Questao 4
# Editando a altura de centimetros para metros de todos da lista
dados_starwars_at["altura"] = dados_starwars_at["altura"] / 100
dados_starwars_at
```

###Questão 5

Renomeie a variável `altura` para `altura_metros`. Salve a base de dados atualizada.

```python
# Questao 5
#alterando nome da coluna
dados_starwars_at = dados_starwars_at.rename(columns={'altura': 'altura_metros'})
dados_starwars_at
```

###Questão 6

Ordene a base de dados por `altura_metros`. Salve a base de dados atualizada.

```python
# Questao 6
# Ordenando a coluna Altura do mais baixo ao mais alto(crescente)
dados_starwars_at = dados_starwars_at.sort_values('altura_metros')
dados_starwars_at
```

###Questão 7

Mantenha na base somente as variáveis: `nome`, `altura_metros`, `massa` e `genero`. Salve a base de dados atualizada.




```python
# Questao 7

novo_starwars = dados_starwars_at[['nome', 'altura_metros', 'massa', 'genero']]
novo_starwars
```
